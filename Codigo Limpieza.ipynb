```python
#Limpieza ESTU_NACIONALIDAD
import re
import pandas as pd

def is_integer(n):
    try:
        float(n)
    except ValueError:
        return False
    else:
        return float(n).is_integer()
    
def clean_text_nationality(text):
    bits = re.split('(\W)', text)
    if is_integer(bits[0]):
        return ''.join(bits[2:])
    else:
        return text
        
datafinal['ESTU_NACIONALIDAD'] = datafinal['ESTU_NACIONALIDAD'].to_frame().applymap(clean_text_nationality)
```
